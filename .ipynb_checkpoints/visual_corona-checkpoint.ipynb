```python
country_dates = dict(Counter(df['location']))
country_dates
```

```python
dates = []
total_cases = []
for i in range(len(df['location'])):
    if df['location'][i] == "United States":
        dates.append(df['date'][i])
        total_cases.append(df['total_cases'][i])

plt.figure(figsize=(country_dates['United States']/5, country_dates['United States']/5))

plt.plot_date(dates, total_cases)
plt.suptitle('USA Total Cases')
plt.show()
```
